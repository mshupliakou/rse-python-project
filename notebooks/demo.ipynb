```python
# 2. Imports
from physcalc.core import simulate_fall
```

## Usage Examples
The `simulate_fall` function uses Numba for acceleration and Pint for unit safety.

```python
# Example 1: Simulation of a heavy object falling for 10 seconds
result_1 = simulate_fall("80 kg", "10 s")
print(f"Example 1 (80 kg, 10 s):")
print(f"Distance: {result_1['distance']}")
print(f"Final Velocity: {result_1['final_velocity']}")
```

```python
# Example 2: A lightweight object falling for a short duration
result_2 = simulate_fall("1 kg", "2 s")
print(f"Example 2 (1 kg, 2 s):")
print(f"Distance: {result_2['distance']}")
print(f"Final Velocity: {result_2['final_velocity']}")
```

```python
# Example 3: Modifying the drag coefficient (e.g., parachute effect)
result_3 = simulate_fall("70 kg", "15 s", drag_coeff=1.2)
print(f"Example 3 (High drag, 70 kg, 15 s):")
print(f"Distance: {result_3['distance']}")
print(f"Final Velocity: {result_3['final_velocity']}")
```
